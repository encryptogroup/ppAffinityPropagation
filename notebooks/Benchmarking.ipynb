```python
random_state = 170


X = np.loadtxt(("../../../test/data/fpcs/lsun.data.gz"), ndmin=2)
Y = np.loadtxt(("../../../test/data/fpcs/lsun.labels0.gz"), ndmin=2)[:,0]

y_pred = AffinityPropagation(preference=-24).fit_predict(X)
model = AffinityPropagation(preference=-24).fit(X)
cluster_centers = model.cluster_centers_

fig, ax = plt.subplots(1, figsize=(4,3))
ax.scatter(X[:, 0], X[:, 1], c=y_pred)
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], color="red")
#plt.title("Affinity Propagation")
ari = sklearn.metrics.adjusted_rand_score(Y, y_pred)
si = sklearn.metrics.silhouette_score(X, y_pred)
print("Affinity Propagation LSUN ARI : ", ari, " SI: ", si)
plt.savefig("lsun_aff_prop.pdf", bbox_inches='tight')


y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)
model = KMeans(n_clusters=3, random_state=random_state).fit(X)
cluster_centers = model.cluster_centers_

fig, ax = plt.subplots(1, figsize=(4,3))
ax.scatter(X[:, 0], X[:, 1], c=y_pred)
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], color="red")
#plt.title("Affinity Propagation")
ari = sklearn.metrics.adjusted_rand_score(Y, y_pred)
si = sklearn.metrics.silhouette_score(X, y_pred)
print("Affinity Propagation LSUN ARI : ", ari, " SI: ", si)
plt.savefig("lsun_kmeans.pdf", bbox_inches='tight')

```

```python
n=100
features=2

D, clust = make_blobs(n_samples=n, centers=6, cluster_std=0.30, random_state=0)#np.load("../../../data/blobs_300.npz")['arr_0']

S = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        for k in range(features):
            S[i][j] = S[i][j] + (D[i][k]-D[j][k])*(D[i][k]-D[j][k])
S = -1*S

preference = np.median(S)
print('preference ',preference)


af = AffinityPropagation(preference=preference, max_iter=10000)
clustering = af.fit(D)
print(clustering.cluster_centers_)
print(clustering.labels_)
print(clustering.n_iter_)
print(clustering.get_params())

fig, ax = plt.subplots(1, figsize=(4,3))

ax.scatter(D[:,0], D[:,1], c=clustering.labels_, alpha=0.7, edgecolors='b')
plt.savefig("blobs_min_6.pdf", bbox_inches='tight')

plt.show()

ari = sklearn.metrics.adjusted_rand_score(clust, clustering.labels_)
si = sklearn.metrics.silhouette_score(D, clustering.labels_)
print("Affinity Propagation LSUN ARI : ", ari, " SI: ", si)
plt.savefig("lsun_aff_prop.pdf", bbox_inches='tight')
```
